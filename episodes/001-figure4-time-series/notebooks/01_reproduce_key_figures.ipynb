# Reproduce key figures

Phase 2 initial target: qualitative reproduction of Fig. 4 trajectories for `T=225 K`, `p=300 hPa`, `F=1`, and `w in {0.01, 0.1, 1} m/s`.

The paper caption does not state initial conditions. The script/notebook uses `x_eq * 0.99`, following the nearby text describing simulations from a small perturbation of equilibrium. This is not a digitized point-by-point validation.


```python
from bergner_spichtinger_2026.constants import Environment
from bergner_spichtinger_2026.core import equilibrium, integrate_trajectory
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(12, 3.2))
for ax, w in zip(axes, [0.01, 0.1, 1.0]):
    env = Environment(p=30000.0, T=225.0, w=w, F=1.0)
    x_eq = equilibrium(env)
    sol = integrate_trajectory(env, x_eq * 0.99, (0, 100_000), max_step=50, rtol=1e-8, atol=1e-13)
    ax.plot(sol.t / 3600, sol.y[2], color='firebrick')
    ax.axhline(x_eq[2], color='black', lw=0.9)
    ax.set_title(f'w={w:g} m/s')
    ax.set_xlabel('time [h]')
    ax.set_ylabel('s')
    ax.grid(alpha=0.25)
fig.tight_layout()
```
